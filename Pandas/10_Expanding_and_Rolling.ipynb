```python
import pandas as pd
import numpy as np
%matplotlib inline
```

```python
s = pd.Series(np.random.randn(1000),
              index=pd.date_range('1/1/2016', periods=1000))
s.head()
```

```python
s = s.cumsum()
s.plot()
```

```python
r = s.rolling(window=30)
r.mean() # window size 만큼 이동시킴, 이동평균(moving average))
```

```python
s.plot(style='k--')
r.mean().plot(style='k')
```

```python
 df = pd.DataFrame(np.random.randn(1000, 4),
                   index=pd.date_range('1/1/2016', periods=1000),
                   columns=['A', 'B', 'C', 'D'])
```

```python
df = df.cumsum()
df.plot()
```

```python
df.rolling(window=60).sum().plot(subplots=True)
```

___

```python
df.rolling(window=len(df), min_periods=1).mean()[:5].plot()
```

```python
df.expanding(min_periods=1).mean().plot()
```
